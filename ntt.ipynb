**4. Pembuatan Model (Modeling)**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Inisialisasi TfidfVectorizer
tfidf = TfidfVectorizer()

ntt['objek'] = ntt['objek'].replace("-", "_", regex=True)
ntt['objek'] = ntt['objek'].replace(" ", "", regex=True)

# Melakukan perhitungan idf pada data genre
tfidf.fit(ntt['objek']) 
 
# Mapping array dari fitur index integer ke fitur nama
tfidf.get_feature_names() 
```

```python
# Melakukan fit lalu ditransformasikan ke bentuk matrix
tfidf_matrix = tfidf.fit_transform(ntt['objek']) 
 
# Melihat ukuran matrix tfidf
tfidf_matrix.shape
```

```python
pd.DataFrame(
    tfidf_matrix.todense(), 
    columns=tfidf.get_feature_names(),
    index=ntt['nama']
).sample(10, axis=1).sample(10, axis=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
 
# Menghitung cosine similarity pada matrix tf-idf
cosine_sim = cosine_similarity(tfidf_matrix) 
cosine_sim
```

```python
# Membuat dataframe dari variabel cosine_sim dengan baris dan kolom berupa nama objek
cosine_sim_df = pd.DataFrame(cosine_sim, index=ntt['nama'], columns=ntt['nama'])
print('Shape:', cosine_sim_df.shape)
 
# Melihat similarity matrix pada setiap objek
cosine_sim_df.sample(5, axis=1).sample(5, axis=0)
```

```python
def ntt_rec_CBF(title_ntt=ntt["nama"], similarity_data=cosine_sim_df, items=ntt[['nama', 'objek']], k=10):
    index = similarity_data.loc[:,title_ntt].to_numpy().argpartition(
        range(-1, -k, -1))
    
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    closest = closest.drop(title_ntt, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```
